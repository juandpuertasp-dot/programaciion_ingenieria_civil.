**Dosificación de Mezcla para Columnas**

**Planteamiento**:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular. Se sabe que para un concreto estándar de 210 kg/cm2
 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

Requerimientos:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v =\pi * r^2 * h$$

. (Sugerencia: usar 3.1416
 o importar math.pi).
Calcular la cantidad de sacos: $$Sacos = v * 9.5$$
.
**Prueba del programa**:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

````python
 # entrada : solicitud de la estatura de la columna.
 # solicitud : el diametro de la columna
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

6. # proceso: calcula el radio r, calcula el volumen de la columna,calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# salida: imprime el resultado.
print(f"Volumen ≈{v:.3f} m3")
print(f"Sacos necesarios ≈{sacos:.1f}")
````

Aquí te presento una explicación línea por línea del código de cálculo de los sacos de cemento:

```python
# entrada : solicitud de la estatura de la columna.
# solicitud : el diametro de la columna
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# proceso: calcula el radio r, calcula el volumen de la columna,calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

print(f"Volumen ≈{v:.3f} m3")
print(f"Sacos necesarios ≈{sacos:.1f}")
```

**Explicación del Código:**

1.  `# entrada : solicitud de la estatura de la columna.`
    *   Este es un comentario que indica la sección de entrada de datos, específicamente para la altura de la columna.

2.  `# solicitud : el diametro de la columna`
    *   Este es otro comentario que indica la entrada de datos para el diámetro de la columna.

3.  `h = float(input("Ingrese la altura de la columna en metros: "))`
    *   Esta línea solicita al usuario que ingrese la altura de la columna en metros. La función `input()` lee la entrada como texto (string). La función `float()` convierte ese texto a un número decimal (flotante) y lo almacena en la variable `h`.

4.  `d = float(input("Ingrese el diámetro de la columna en metros: "))`
    *   Similar a la línea anterior, esta línea solicita al usuario el diámetro de la columna en metros, lo convierte a un número flotante y lo guarda en la variable `d`.

5.  `# proceso: calcula el radio r, calcula el volumen de la columna,calcula sacos`
    *   Este comentario delimita la sección donde se realizan los cálculos.

6.  `r = d/2`
    *   Calcula el radio de la columna dividiendo el diámetro (`d`) por 2 y asigna el resultado a la variable `r`.

7.  `v = 3.1416 * r**2 * h`
    *   Calcula el volumen de la columna cilíndrica utilizando la fórmula `V = π * r² * h`. Aquí, `3.1416` se usa como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y `h` es la altura. El resultado se guarda en la variable `v`.

8.  `sacos = v * 9.5`
    *   Calcula la cantidad de sacos de cemento necesarios multiplicando el volumen de la columna (`v`) por la cantidad de sacos requerida por metro cúbico (9.5). El resultado se almacena en la variable `sacos`.

9.  `# salida: imprime el resultado.`
    *   Este comentario indica la sección de salida de los resultados.

10. `print(f"Volumen ≈{v:.3f} m3")`
    *   Imprime el volumen calculado (`v`) formateado para mostrar solo tres decimales (`:.3f`). La `f` antes de la cadena indica un *f-string*, que permite incrustar variables directamente en la cadena.

11. `print(f"Sacos necesarios ≈{sacos:.1f}")`
    *   Imprime la cantidad de sacos de cemento calculada (`sacos`) formateada para mostrar solo un decimal (`:.1f`).
