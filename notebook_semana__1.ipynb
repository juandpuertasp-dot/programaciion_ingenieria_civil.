```python
# entrada: ancho y alto en metros
ancho = float(input("ingrese ancho de la ventana en metros"))
alto = float(input("ingrese el alto de la ventana en metros"))

#proceso calcula lm en pies y av en metros cuadrados
lm = 2 * (ancho + alto) * 3.25
av = (ancho * alto) * 2

#salida: sale longitud de la madera (lm) en pies y areas del vidrio en metros cuad
print(f"longitud de la madera: {lm} pies de longitud")
print(f"area del vidrio: {av} metros cuadrado")
```
